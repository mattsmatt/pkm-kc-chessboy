```python
import engine as ce
import chess as ch
import chess.svg
from PyQt5.QtSvg import QSvgWidget
from PyQt5.QtWidgets import QApplication, QWidget
from IPython.display import clear_output

class Main:

    def __init__(self, board=ch.Board):
        self.board=board

    #play human move
    def playHumanMove(self):
        try:
            display(self.board)
            print(self.board.legal_moves)
            print("""To undo your last move, type "undo".""")
            #get human move
            play = input("Your move: ")
            clear_output()
            if (play=="undo"):
                self.board.pop()
                self.board.pop()
                self.playHumanMove()
                return
            self.board.push_san(play)
        except:
            self.playHumanMove()

    #play engine move
    def playEngineMove(self, maxDepth, color):
        engine = ce.Engine(self.board, maxDepth, color)
        self.board.push(engine.getBestMove())

    #start a game
    def startGame(self):
        color=None
        while(color!="b" and color!="w"):
            color = input("""Play as (type "b" or "w"): """)
        maxDepth=None
        while(isinstance(maxDepth, int)==False):
            maxDepth = int(input("""Choose depth: """))
        if color=="b":
            while (self.board.is_checkmate()==False):
                self.playEngineMove(maxDepth, ch.WHITE)
                self.playHumanMove()
            # print(self.board.outcome())
            outcome = self.board.outcome()
            if outcome:
                if outcome.winner == chess.WHITE:
                    print("white won")
                elif outcome.winner == chess.BLACK:
                    print("black won")
                else:
                    print("draw")
        elif color=="w":
            while (self.board.is_checkmate()==False):
                self.playHumanMove()
                self.playEngineMove(maxDepth, ch.BLACK)
            # print(self.board.outcome())
            outcome = self.board.outcome()
            if outcome:
                if outcome.winner == chess.WHITE:
                    print("white won")
                elif outcome.winner == chess.BLACK:
                    print("black won")
                else:
                    print("draw")
        self.board.reset
        self.startGame()

#create an instance and start a game

newBoard = ch.Board()
game = Main(newBoard)
cherryBoy = game.startGame()
```
